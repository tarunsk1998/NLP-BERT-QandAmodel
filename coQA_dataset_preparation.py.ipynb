```python
import pandas as pd
import numpy as np
```

```python
input_df = pd.read_json('http://downloads.cs.stanford.edu/nlp/data/coqa/coqa-train-v1.0.json')
input_df.head()
```

```python
input_df.loc[0]["data"]
```

```python
input_df = input_df.join(pd.json_normalize(input_df.data))
```

```python
cols = ["story","question","answer"]
coqa_list = []
for i in range(len(input_df)):
    for j in range(len(input_df.iloc[i]["questions"])):
        #print(i, input_df.iloc[i]["questions"][j]["input_text"], input_df.iloc[i]["answers"][j]["input_text"])
        coqa_list.append([input_df.iloc[i]["story"], input_df.iloc[i]["questions"][j]["input_text"], 
                                 input_df.iloc[i]["answers"][j]["input_text"]])
coqa = pd.DataFrame(coqa_list, columns = cols)
```

```python
coqa.head()
```

```python
coqa.to_csv("CoQA_data.csv", index=False)
```
